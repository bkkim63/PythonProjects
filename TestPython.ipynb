### rrange() function
- range() 함수의 변형
- 정수 및 실수 시퀀스 생성이 가능하며, end를 포함하도록 구현한 것
    - 단, end는 step에 따라 포함되지 않을 수 있음

```python
######################################################################################
# range() 함수의 변형(정수 및 실수 시퀀스 생성이 가능하며, end를 포함하도록 구현한 것)
# name : real range function (rrange)
#---------------------------------------------------- by B.K.Kim 2024-06-17 ----------
def rrange(*args):
    # 1. 인수 개수 검증
    if len(args) == 0 or len(args) > 3:
        raise TypeError(f"rrange expected 1 to 3 arguments, got {len(args)}")
    # 2. 인수 파싱
    if len(args) == 1:
        start, end, step = 0, args[0], 1
    elif len(args) == 2:
        start, end, step = args[0], args[1], 1
    else:
        if args[2].is_integer():
            start, end, step = args[0], args[1], args[2]
        else:
            start, end, step = float(args[0]), float(args[1]), float(args[2])
    # 3. step 값 검증
    if step == 0:
        raise ValueError("rrange() arg 3 must not be zero")
    # 4. 범위에 따른 시퀀스 생성
    if step > 0:
        i = start
        while i <= end:
            yield round(i, 15)
            i += step
    else:  # step < 0
        i = start
        while i >= end:
            yield round(i, 15)
            i += step

if __name__ == "__main__":
    print(list(rrange(5)))
    print(list(rrange(1, 5)))
    print(list(rrange(0, 10, 2)))
    print(list(rrange(10, 0, -2)))
    print(list(rrange(0, 1, 0.2)))
    print(list(rrange(0, 1, 0.3)))
    print(list(rrange(1, 0, -0.1)))
```

### Object-Oriented Programming

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def set(self, x, y):
        self.x = x
        self.y = y
    def showPoint(self):
        print(f"({self.x}, {self.y})")

class ColorPoint(Point):
    def __init__(self, x, y, color):
        super().__init__(x, y)
        self.color = color
    def setColor(self, color):
        self.color = color
    def showColorPoint(self):
        print(f"({self.x}, {self.y}, {self.color})")

if __name__ == "__main__":
    p = Point(1, 2)
    p.showPoint()
    cp = ColorPoint(3, 4, "red")
    cp.showColorPoint()
```

```python
class Shape:
    def draw(self):
        print(f"Drawing a shape")

class Line(Shape):
    def draw(self):
        print(f"Drawing a line")

class Rect(Shape):
    def draw(self):
        print(f"Drawing a rect")

class Circle(Shape):
    def draw(self):
        print(f"Drawing a circle")

def paint(shape):
    shape.draw()
    
if __name__ == "__main__":
    paint(Shape())
    paint(Line())
    paint(Rect())
    paint(Circle())
```

```python
class ParentClass:
    role = "Dad"

    def __init__(self, data):
        self.data = data

    @classmethod
    def class_method_role(cls):
        print(cls.role)

    @staticmethod
    def static_method_role():
        print(ParentClass.role)


class ChildClass(ParentClass):
    role = "Son"

if __name__ == "__main__":
    ChildClass.class_method_role()
    ChildClass.static_method_role()
```
